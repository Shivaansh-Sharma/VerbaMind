```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import joblib
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score,classification_report
from sklearn.preprocessing import LabelEncoder
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
le=LabelEncoder()
```

```python
data=pd.read_csv(r"C:\Users\Karan Brar\Desktop\New folder\Sentiment_Analyzer\data\new_train.csv")
```

```python
text=data['text']
text=text.fillna("")
y=data['sentiment']
y=y.dropna()
```

```python
print(text.count())
print(y.count())
```

```python
vectorizer=TfidfVectorizer(analyzer='char',ngram_range=(1,5))
X=vectorizer.fit_transform(text)
```

```python
y_encoded=le.fit_transform(y)
```

```python
model=MultinomialNB()
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y_encoded,train_size=0.8)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
print(y_pred)
```

```python
print(accuracy_score(y_pred,y_test))
```

```python
from xgboost import XGBClassifier

clf = XGBClassifier(
    objective="multi:softmax",    # outputs class label directly
    num_class=3,                  # set number of classes explicitly (e.g., 3 classes)
    eval_metric="mlogloss",       # good for multi-class classification
    use_label_encoder=False,
    random_state=42,
    
    # 🔑 Hyperparameters to improve accuracy
    n_estimators=500,             # more boosting rounds
    learning_rate=0.05,           # smaller step size for better generalization
    max_depth=6,                  # controls complexity of trees
    subsample=0.8,                # use 80% of data for each boosting round
    colsample_bytree=0.8,         # use 80% of features for each tree
    reg_lambda=1.0,               # L2 regularization (helps avoid overfitting)
    gamma=1,                      # minimum loss reduction (makes splits more conservative)
    min_child_weight=3            # minimum sum of instance weights (avoids overfitting small samples)
)
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
print(accuracy_score(y_pred,y_test))
```
